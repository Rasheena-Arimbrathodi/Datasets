4. IRIS dataset has four features: sepal length, sepal width, petal length, and petal width that can be
used to identify the kind of IRIS species.
1
(a) Perform PCA on the given dataset and reduce the dimension to two
(b) Plot the newly obtained features (Principal components) discriminating the species type
(c) Identify which one of the four features can be used to discriminate between the species (Use
the coefficient of principal components to answer)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Iris - Iris.csv")
```

```python
df=pd.get_dummies(df, columns = ['Species'])
```

```python
df.head(5)
```

```python
U,S,Vt=np.linalg.svd(df)
```

```python
print(U.shape,S.shape,Vt.shape)
```

```python
# (a)
def get_princ_comp(df,Vt,n):
   '''
   df:data matrix
   Vt:right singular matrix
   '''
   pc=np.dot(df,Vt.T)
   return pc[:,:n]
```

```python
# (b)
PC2=get_princ_comp(df,Vt,2)
plt.plot(PC2)
```

```python
print(PC2)
```

```python
#(c)
PC1=get_princ_comp(df,Vt,1)
print(PC1)
```

```python
# Assuming you have obtained the first principal component as PC1

# Get the names of your original variables (assuming you have them in a list)
variable_names = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]  # Replace with your variable names

# Create a list of (variable name, loading value) pairs
variable_loadings = list(zip(variable_names, PC1))

# Sort the variables based on the absolute loading values in descending order
sorted_variables = sorted(variable_loadings, key=lambda x: abs(x[1]), reverse=True)

# Print the sorted variables and their loadings
for variable, loading in sorted_variables:
    loading_value = loading.item()  # Extract the scalar value from the NumPy array
    print(f"{variable}: {loading_value:.4f}")

```

# PetalWidthCm has the highest absolute loading value of approximately -4.4334.
The first principal component (PC1) corresponds to the linear combination of these original variables. In this case, PetalWidthCm is the variable that is most strongly associated with PC1, (*followed by PetalLengthCm, SepalWidthCm, and SepalLengthCm*).The negative sign of these loading values indicates the direction of the association. Variables with negative loading values are inversely related to PC1, meaning that as the values of these variables increase, the value of PC1 decreases. Conversely, as the values of PetalWidthCm and the other positively loaded variables increase, the value of PC1 increases.

So, in summary, PetalWidthCm is the primary variable most strongly associated with the first principal component, and it contributes the most to the variation explained by PC1 in the given dataset.
