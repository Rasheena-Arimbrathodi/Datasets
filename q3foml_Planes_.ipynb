```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy.random.mtrand import RandomState
X=np.random.normal(0,2,size=(200,10))
```

```python
plt.figure(figsize=(20,10))
_,bins,_=plt.hist(X,30,density=True)
fx=1/(2*np.sqrt(2*np.pi))*np.exp(-(bins-0)**2/(2*2**2))
plt.plot(bins,fx,linewidth=3,color='b')
plt.show()
```

```python
#SVD
U,S,Vt=np.linalg.svd(X)
```

```python
print(U.shape,S.shape,Vt.shape)
```

```python
def get_principal_component(X,Vt,n):
  '''X: data matrix
    Vt: right singular matrix
  '''
  PC=X@(Vt.T)
  return(PC[:,:n])
```

```python
PC=get_principal_component(X,Vt,2)
print(PC)
```

```python
def get_var_exp(S,n):
  '''S:vectors of singular values
  '''
  eig=S**2
  return eig[n]/np.sum(eig)
```

```python
# (a)  First principal component
first_comp_var_exp=get_var_exp(S,1)
print(first_comp_var_exp)
```

```python
# Second principal component
second_comp_var_exp=get_var_exp(S,2)
print(second_comp_var_exp)
```

```python
M=list(range(1,10)) #components
var_exp=[get_var_exp(S,i-1)for i in M]
```

```python
plt.plot(M,var_exp,"yd-")
plt.xlabel("Components")
plt.ylabel("Variance explained")
plt.title("Scree Plot")
```

```python
#function to reconstruct the data from the principal components
def recons_data_mat(PC,Vt,n):
  """PC: principal components
  Vt: right singular matrix
  """
  X_=PC@(Vt[:n,:])
  return X_
```

```python
X_=recons_data_mat(PC,Vt,2)
print(X_)
```

```python
X_.shape
```

```python
# b)
#calculating the SSE for the elements at each index.
SSE = sum((X - X_)**2)
print(SSE)
```

```python
sse=(sum(SSE))
print(sse)
```

```python
# (c) Function to generate a random plane
import random
def generate_random_plane():
    x = random.choice(X)
    y = random.choice(X)
    return {
        "x": x,
        "y": y,
    }

# Generate 50 random planes
random_planes = [generate_random_plane() for _ in range(50)]

# Print the generated planes
for idx, plane in enumerate(random_planes, start=1):
    print(f"Plane {idx}:")
    for key, value in plane.items():
        print(f"{key}: {value}")

    print()
```

```python
import random

# Let's make a list to store these 50 planes
planes = []

# Create 50 random planes
for _ in range(50):
    # Generate random numerical attributes for each plane
    attributes = [random.randint(1, 100) for _ in range(8)]  # Let's assume 8 numerical attributes
    planes.append(attributes)
```

```python
# Create an empty grid (matrix) with 200 rows and 10 columns
grid = [[None for _ in range(10)] for _ in range(200)]

# Fill the grid by repeating the 50 planes
for i in range(200):
    for j in range(10):
        # Use % to loop through the 50 planes in a circular way
        plane = planes[(i * 10 + j) % 50]
        grid[i][j] = plane
```

```python
# Calculate the total sum of numerical attributes for each plane and store it in a list
total_sums = [sum(plane) for plane in planes]
print(total_sums)
```
