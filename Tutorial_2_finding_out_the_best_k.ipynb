```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("housing.csv",usecols=['latitude',"longitude",'median_house_value'])
df.head()
```

## Normalising the data
1. When working with distance based algorithms, like k-means clustering , we must normalise the data.
2. We first set up training and test splits using train_test_split from sklearn.
3. 33% of the data will be used for testing and 67% will be used for training.
4. Random_state parameters is set to 0 to ensure that the same random split is generated each time the code is run.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[['latitude','longitude']],df[['median_house_value']],test_size=0.33,random_state=0)
```

We normalise the training and test data using the preprocessing. normalise() method from sklearn.This function scales the input data so that each sample (i.e.,row) has a unit norm

```python
from sklearn import preprocessing
X_train_norm=preprocessing.normalize(X_train)
X_test_norm=preprocessing.normalize(X_test)
```

## Fitting and Evaluating the model
1. We will arbitrarily choose a number of clusters(referred to as k) of 3
2. We will create an instance of KMeans, defines the number of clusters using the n clusters attribute.
3. We can then fit the model to the normalised training data using the fit() method.


```python
from sklearn.cluster import KMeans
kmeans= KMeans(n_clusters=2,random_state=0,n_init="auto")
kmeans.fit(X_train_norm)
```

```python
sns.scatterplot(data=X_train, x='longitude',y='latitude',hue=kmeans.labels_)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(X_train_norm,kmeans.labels_,metric="euclidean")
```

```python
silhouette_score(X_train_norm,kmeans1.labels_,metric="euclidean")
```
